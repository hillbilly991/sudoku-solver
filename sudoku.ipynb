## Part One
You should write all of your code for solving sudokus below this cell.

You must include a function called `sudoku_solver(sudoku)` which takes one sudoku puzzle (a 9x9 NumPy array) as input, and returns the solved sudoku as another 9x9 NumPy array. This is the function which will be tested. 

```python
class PartialSudokuState:
    def __init__(self, board):
        self.board = board
        self.check_row = [[False for _ in range(10)] for _ in range(9)]
        self.check_column = [[False for _ in range(10)] for _ in range(9)]
        self.check_block = [[False for _ in range(10)] for _ in range(9)]

        # This loops through each of the values in the board, sees whether any are
        # filled in (not 0), and then adds these to the check_row, check_column and
        # check_block arrays respectively
        # The purpose of using these is to make sure that we can always keep track
        # of which values are already taken in each row, column or block
        for row in range(9):
            for col in range(9):
                if board[row][col] != 0:
                    # It might seem odd that I've opted to use the actual number value
                    # for indexes, but it means that I'm not needing to do +1 everywhere
                    # when I want to check whether the value has been used or not. Also,
                    # the convenience of doing it this way outweighs the possible space
                    # complexity implications since it's only adding one extra index to
                    # each array
                    self.check_row[row][board[row][col]] = True
                    self.check_column[col][board[row][col]] = True
                    block_id = self.get_block_id(row, col)
                    self.check_block[block_id][board[row][col]] = True

    def can_place_value_in_cell(self, row, col, value):
        return (
            self.__can_place_value_in_row__(row, value)
            and self.__can_place_value_in_column__(col, value)
            and self.__can_place_value_in_block__(row, col, value)
        )

    def is_goal_state(self):
        return all(
            (
                self.__board_has_no_zeroes__(),
                self.__all_values_in_rows_are_unique__(),
                self.__all_values_in_columns_are_unique__(),
                self.__all_values_in_blocks_are_unique__(),
            )
        )

    def get_block_id(self, row, col):
        """
        A block is a 3x3 square starting from the top left of the sudoku board to the
        bottom right of the sudoku board. To represent this, there is a list with 9
        indexes; each list represents one of the blocks and each index in the list
        (which is 9 indexes long) represents whether the value is already in the
        block

        For example, if 4 was in the top left block you would see True in the first list and
        the fourth index like so:
        [
          [False, False, False, True, False, False, False, False, False, False],
          [False, ....],
          ....
        ]

        n.b: it's the fourth index (not third) for convenience - otherwise I'd often be doing
        index + 1

        The blocks 'wrap around' the board, so a value in the fourth block would be a value
        within the first three columns of the second row (since there are three blocks in the
        first row)

        This is the reason for the below calculation since it mimics this wrap around effect
        that happens on the board.

        1. overall it puts rows into three bands; 0-2, 3-5, and 6-8
        2. (row // 3) * 3 returns a value that is either 0, 3 or 6 (the beginning of each row)
        2. (col // 3) then pushes the value of (row // 3) further to the right and causes a
           wrap around if necessary
        """
        return (row // 3) * 3 + (col // 3)

    def final_state(self):
        if not self.is_goal_state():
            return np.full((9, 9), -1)

        return np.array(self.board)

    def find_least_constraining_values_cell(self):
        """
        The least constraining values cell is the cell that has the least number of possible numbers
        that can be added
        """
        minimum_number_of_options = 10
        least_constraining_values_cell = None
        possible_values_for_least_constraining_values_cell = []

        for row in range(9):
            for col in range(9):
                if self.board[row][col] != 0:
                    continue

                potential_values = [
                    value
                    for value in range(1, 10)
                    if self.can_place_value_in_cell(row, col, value)
                ]

                if len(potential_values) < minimum_number_of_options:
                    minimum_number_of_options = len(potential_values)
                    least_constraining_values_cell = (row, col)
                    possible_values_for_least_constraining_values_cell = (
                        potential_values
                    )

        return (
            least_constraining_values_cell,
            possible_values_for_least_constraining_values_cell,
        )

    def __can_place_value_in_row__(self, row, value):
        if self.check_row[row][value]:
            return False

        return True

    def __can_place_value_in_column__(self, col, value):
        if self.check_column[col][value]:
            return False

        return True

    def __can_place_value_in_block__(self, row, col, value):
        block_id = self.get_block_id(row, col)

        if self.check_block[block_id][value]:
            return False

        return True

    def __board_has_no_zeroes__(self):
        return np.all(self.board != 0)

    def __all_values_in_rows_are_unique__(self):
        for row in self.board:
            if np.unique(row).size != 9:
                return False

        return True

    def __all_values_in_columns_are_unique__(self):
        for col in range(9):
            if np.unique(self.board[:, col]).size != 9:
                return False

        return True

    def __all_values_in_blocks_are_unique__(self):
        for x in range(3):
            for y in range(3):
                start_x, end_x = [3 * x, 3 * x + 3]
                start_y, end_y = [3 * y, 3 * y + 3]
                if np.unique(self.board[start_x:end_x, start_y:end_y]).size != 9:
                    return False

        return True


def depth_first_search(partial_sudoku_state):
    (
        least_constraining_values_cell,
        possible_values_for_least_constraining_values_cell,
    ) = partial_sudoku_state.find_least_constraining_values_cell()
    # If there is no other cell to choose from then there are no options and therefore
    # the board has been filled
    if least_constraining_values_cell is None:
        return True

    row, col = least_constraining_values_cell
    block_id = partial_sudoku_state.get_block_id(row, col)
    for value in possible_values_for_least_constraining_values_cell:
        partial_sudoku_state.board[row][col] = value
        partial_sudoku_state.check_row[row][value] = True
        partial_sudoku_state.check_column[col][value] = True
        partial_sudoku_state.check_block[block_id][value] = True
        # If we depth first search and it is true all the way up the call stack
        # then we've found a solution
        if depth_first_search(partial_sudoku_state):
            return True
        # If depth first search returned false then we know we have not found a
        # a solution from the above depth first search branch, so we backtrack
        # and reset everything in the constraint lists to false below
        else:
            partial_sudoku_state.check_row[row][value] = False
            partial_sudoku_state.check_column[col][value] = False
            partial_sudoku_state.check_block[block_id][value] = False

    partial_sudoku_state.board[row][col] = 0

    return False


def sudoku_solver(sudoku):
    partial_sudoku_state = PartialSudokuState(sudoku)
    depth_first_search(partial_sudoku_state)

    return partial_sudoku_state.final_state()
```
